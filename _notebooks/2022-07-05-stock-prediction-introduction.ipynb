# Stock Prediction Project 

- __주제__
    - 주가 데이터셋을 사용하여 __`인공지능 기반 주가 예측 모델`__을 연구한다.
- __주가 데이터의 장점__ 
    - 주가 데이터는 접근성이 좋으면서도 매일 방대한 양이 쏟아지는 빅데이터이다. 데이터 수집이 어렵지 않고, 데이터의 양도 많기 때문에 데이터를 다루는 훈련을 하기에 아주 적합하다.
    - 일상생활과 밀접해있어 보다 `실용적인` 빅데이터 연구가 가능하다. 
    -  주식은 사회경제적 관심과 영향력이 크기 때문에 주가 데이터를 활용한 연구는 학계에서도 활발히 진행중이다. 
- __프로젝트 진행 요약__
    - `baseline model selection`(2022.01): 주가 예측에 대한 문제 정의 및 그에 맞는 데이터셋을 구성한다. 기본적인 데이터 전처리와 모델링을 진행한 후, 가장 좋은 성능을 보이는 머신러닝 모델을 선정한다. 최종적으로 자동매매 시뮬레이션을 통해 수익률을 구한다. 
    - `논문 연구`(2022.02~): 데이터를 더 자세히 들여다 보고, feature 추가, 데이터 필터링 등 데이터의 질을 향상시킨다. 설명가능 AI(XAI)기법 중 하나인 SHAP value를 사용하여 모델에 대한 해석을 중점으로 두어 연구하고, 논문주제를 발굴하여 논문 작성을 진행한다.
