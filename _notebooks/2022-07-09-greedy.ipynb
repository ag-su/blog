# 거스름 돈 

```python
lst_coin = [500, 100, 50, 10]
money = 1260
coin_cnt = 0
for coin in lst_coin: 
    change = money % coin
    if change < 0: 
        continue
    coin_cnt += (money // coin)
    money = change
print(coin_cnt)
```

# 1이 될 때 까지 

```python
# "최소" -> 그리디 문제임을 의심 해보자. 
n, k = map(int, input().split())

cnt = 0 
while n > 1: 
    if n % k == 0: 
        n //= k
        cnt += 1
    else:
        n -= 1
        cnt += 1
print(cnt)
```

# prac 문제 1

```python
def maxProfit(prices):
    cur = prices[0]
    total = 0
    for i in range(1, len(prices)):
        if cur <= prices[i]:
            total += prices[i] - cur
        cur = prices[i]
    return total

# 아래는 수정하지 마시오.
prices01 = [7,1,5,3,6,4]
print(maxProfit(prices01))
```

# prac 문제 2

```python
def findContentChildren(g, s):
    g.sort()
    s.sort()

    i = 0  # 아이 인덱스  
    j = 0  # 쿠키 인덱스
    
    while (i < len(g)) and (j < len(s)): 
        if g[i] <= s[j]: # 만족 시킴 
            i += 1 
        j += 1
        
        
    return i
        
                

# 아래는 수정하지 마시오.
print(findContentChildren([1,2,3], [1,1]))
print(findContentChildren([1,2], [1,2,3]))
print(findContentChildren([2,3,4], [1,5,3,1]))
```

# prac 문제 2 

```python
def findContentChildren(g, s): 
    g.sort()
    s.sort()
    
    i = 0 # 아이의 인덱스 
    j = 0 # 쿠키의 인덱스 
    
    while i < len(g) and j < len(s): 
        # 아이의 만족하는 쿠키 크기 <= 쿠기 크기
        # -> 만족 카운트 
        if g[i] <= s[j]:
            i += 1 
            j += 1
        # 그렇지 않으면 -> 다음 쿠키로
        else: 
            j += 1
    return i
         
    
# 아래는 수정하지 마시오.
print(findContentChildren([1,2,3], [1,1]))
print(findContentChildren([1,2], [1,2,3]))
print(findContentChildren([2,3,4], [1,5,3,1]))    
```
